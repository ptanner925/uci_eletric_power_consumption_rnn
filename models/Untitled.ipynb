```python
import os, tensorflow as tf, numpy as np
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"  # CPU only sanity check

tf.config.run_functions_eagerly(True)      # easier debugging
X = np.random.randn(2048, 32).astype("float32")
y = np.random.randn(2048, 1).astype("float32")

m = tf.keras.Sequential([tf.keras.layers.Input((32,)),
                         tf.keras.layers.Dense(64, activation="relu"),
                         tf.keras.layers.Dense(1)])
m.compile(optimizer="adam", loss="mse")
m.fit(X, y, epochs=1, batch_size=32, verbose=2)
print("CPU sanity run ✓")
```

```python
import os, time, tensorflow as tf, numpy as np
os.environ.pop("CUDA_VISIBLE_DEVICES", None)  # allow GPU/Metal

print("Devices:", tf.config.list_physical_devices())
tf.config.run_functions_eagerly(False)        # normal graph mode

# Tiny matmul timing to confirm GPU path is alive
a = tf.random.normal([4096, 4096])
b = tf.random.normal([4096, 4096])

t0 = time.time(); c = tf.matmul(a, b); _ = c.numpy(); t1 = time.time()
print(f"Matmul wall time: {t1 - t0:.3f}s  (GPU should be clearly faster than CPU sanity test)")
```
